```python
import torch
import matplotlib.pyplot as plt

torch.set_grad_enabled(False)
```

```python
t = torch.linspace(0, 1, 8)

t1, t2 = torch.meshgrid(t, t, indexing='xy')
```

```python
t1t2 = torch.stack((t1, t2))
```

```python
t1t2.shape
```

```python
t1t2_flat = t1t2.reshape(2, -1)
```

```python
A = torch.tensor([[2, 0], [0, 2]]).float()
```

```python
A.shape, t1t2_flat.shape
```

```python
r = A@t1t2_flat
```

```python
r.shape
```

```python
plt.matshow(r)
```

```python
plt.quiver(t1, t2, r[0], r[1], torch.linalg.norm(r, dim=0))
plt.show()
```

```python
s = r.T @ A
```

```python
s.shape
```

```python
plt.matshow(s)
```
