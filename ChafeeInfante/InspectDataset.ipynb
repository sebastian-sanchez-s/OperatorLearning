```python
import torch
import matplotlib.pyplot as plt

torch.set_grad_enabled(False)
```

```python
path = 'dataset/ds10k/train/'
train_ds_size = 10000

for i in range(0, train_ds_size, train_ds_size//10):
    x = torch.load(path + f'{i}_X.pt')
    y = torch.load(path + f'{i}_Y.pt')
    xprime = (torch.roll(x,-1) - x) * 200

    fig, ax = plt.subplots(ncols=3, figsize=(9, 3))
    fig.tight_layout()

    ax[0].plot(x)
    ax[0].set_xlabel('t')
    ax[1].plot(y)
    ax[1].set_xlabel('t')
    ax[2].plot(x, xprime)

    plt.show()
    plt.close()
```
