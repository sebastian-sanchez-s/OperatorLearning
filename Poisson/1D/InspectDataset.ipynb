```python
import numpy as np
import torch
import matplotlib.pyplot as plt
```

```python
size = 8_000
path = 'dataset/ds10k/train/'

data = []

for i in range(0, size):
    x = torch.load(path + f'{i}_X.pt').detach()
    y = torch.load(path + f'{i}_Y.pt').detach()
    data.append((x,y))
```

```python
for i in range(0, size, size//10):
    fig, ax = plt.subplots(ncols=2)
    fig.tight_layout()

    x, y = data[i]
    
    ax[0].plot(x)
    ax[0].set_xlabel('t')
    ax[0].set_ylabel('x(t)')
    
    ax[1].plot(y)
    ax[1].set_xlabel('t')
    ax[1].set_ylabel('y(t)')
    
    plt.show()
```
