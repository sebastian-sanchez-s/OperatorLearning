```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#  Basis Model
valid_err = []
train_err = []
with open('dataset/ds20k/model/error_ds20k_basis.log') as f:
    for line in f.readlines():
        columns = line.split()
        if int(columns[1]) % 10 == 0:
            valid_err.append(float(columns[7]))
            train_err.append(float(columns[13]))
```

```python
plt.figure(figsize=(5, 5))
plt.loglog(valid_err, label='valid')
plt.loglog(train_err, label='train')
plt.xlabel('every 10 epochs')
plt.legend()
plt.savefig('figures/basis_errors.png', bbox_inches='tight')
plt.show()
```
