```python
import numpy as np
import torch
import matplotlib.pyplot as plt
```

```python
size = 100
path = 'dataset/test/valid/'

data = []

for i in range(0, size):
    x = torch.load(path + f'{i}_X.pt').detach()
    y = torch.load(path + f'{i}_Y.pt').detach()
    data.append((x,y))
```

```python
for i in range(0, size, size//10):
    fig, ax = plt.subplots(ncols=2)
    fig.tight_layout()

    x, y = data[i]
    
    p1 = ax[0].imshow(x, cmap='turbo')
    ax[0].set_xlabel('t')
    ax[0].set_ylabel('t')
    plt.colorbar(p1, ax=ax[0])
    
    p2 = ax[1].imshow(y.view(100, 100), cmap='turbo')
    ax[1].set_xlabel('t')
    ax[1].set_ylabel('t')

    plt.colorbar(p2, ax=ax[1])
    plt.show()
```
